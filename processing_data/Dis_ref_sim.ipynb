```python
import sys
import numpy as np
import xarray as xr
import geopandas as gpd
import torch as torch
import pandas as pd

# Define the file path on Argon
file_path = "/net/argon/landclim/fcourvoisier/runoff_projection/data/raw/Discharge_reference_simulations.txt"

# Load the file with space as separator and handle quoted column names
df = pd.read_csv(file_path, sep=" ", header=0, quotechar='"', engine="python")

# Convert 'Date' column to datetime and set as index
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

# Remove quotes and extra spaces from catchment names
df.columns = df.columns.str.replace('"', '').str.strip()

# Convert DataFrame to Xarray Dataset (Catchment as a coordinate)
ds = xr.Dataset(
    {"Discharge": (["Date", "Catchment"], df.values)},
    coords={"Date": df.index, "Catchment": df.columns}
)

# Display dataset structure
print(ds)
ds
```
